```python
def calculate_final_price():
    price = int(input("Enter price: "))
    dis_rate = int(input("Enter discount (%): "))

    discount_amount = (price * dis_rate) / 100
    final_price = price - discount_amount

    print(final_price)

calculate_final_price()
```
