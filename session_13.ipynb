```python
playlist = ["bewafa", "believer", "despacito", "thug_le", "adat_se_majboor"]

def print_playlist_songs(index):
    if index == len(playlist):
        return

    print(playlist[index])
    print_playlist_songs(index + 1)

print_playlist_songs(0)



def count_unread_messages(messages):
    # Count unread messages in the current group
    total = messages.get("count", 0)

    # Recursively count unread messages in subgroups
    for subgroup in messages.get("subgroups", []):
        total += count_unread_messages(subgroup)

    return total


# Example nested dictionary
whatsapp_groups = {
    "name": "College",
    "count": 5,
    "subgroups": [
        {
            "name": "Project Team",
            "count": 3,
            "subgroups": []
        },
        {
            "name": "Sports Club",
            "count": 2,
            "subgroups": [
                {
                    "name": "Cricket Team",
                    "count": 4,
                    "subgroups": []
                },
                {
                    "name": "Football Team",
                    "count": 1,
                    "subgroups": []
                }
            ]
        }
    ]
}

# Function call
print("Total unread messages:", count_unread_messages(whatsapp_groups))


# q-3
''''x = 'global' :-this is global
def outer():
x = 'outer'    :-this is local
def inner():
nonlocal x
x = 'inner'
inner()
print('Inside outer:', x)
outer()                 :-when this running it  gives global as answer.
print('Outside:', x)'''


def format_number_short(n):
    # Base case
    if n < 1000:
        return str(n)

    # Thousands
    if n < 1000000:
        return str(round(n / 1000, 1)) + "K"

    # Millions (recursive call)
    return format_number_short(n / 1000).replace("K", "M")


# Test cases
print(format_number_short(500))        # 500
print(format_number_short(1500))       # 1.5K
print(format_number_short(25000))      # 25.0K
print(format_number_short(1200000))    # 1.2M
print(format_number_short(3500000))    # 3.5M
```
